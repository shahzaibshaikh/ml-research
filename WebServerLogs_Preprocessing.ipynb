```python
common_regex = '^(?P<client>\S+) \S+ (?P<userid>\S+) \[(?P<datetime>[^\]]+)\] "(?P<method>[A-Z]+) (?P<request>[^ "]+)? HTTP/[0-9.]+" (?P<status>[0-9]{3}) (?P<size>[0-9]+|-)'
combined_regex = '^(?P<client>\S+) \S+ (?P<userid>\S+) \[(?P<datetime>[^\]]+)\] "(?P<method>[A-Z]+) (?P<request>[^ "]+)? HTTP/[0-9.]+" (?P<status>[0-9]{3}) (?P<size>[0-9]+|-) "(?P<referrer>[^"]*)" "(?P<useragent>[^"]*)'
columns = ['client', 'userid', 'datetime', 'method', 'request', 'status', 'size', 'referer', 'user_agent']
```

```python
import time
import re
import pandas as pd


def logs_to_df(logfile, output_dir, errors_file):
    with open(logfile) as source_file:
        linenumber = 0
        parsed_lines = []
        for line in tqdm(source_file):
            try:
                log_line = re.findall(combined_regex, line)[0]
                parsed_lines.append(log_line)
            except Exception as e:
                with open(errors_file, 'at') as errfile:
                    print((line, str(e)), file=errfile)
                continue
            linenumber += 1
            if linenumber % 250_000 == 0:
                df = pd.DataFrame(parsed_lines, columns=columns)
                df.to_parquet(f'{output_dir}/file_{linenumber}.parquet')
                parsed_lines.clear()
        else:
            df = pd.DataFrame(parsed_lines, columns=columns)
            df.to_parquet(f'{output_dir}/file_{linenumber}.parquet')
            parsed_lines.clear()
```

```python
# Define a function to generate request route templates
# Define a function to generate request route templates
def modify_request(route):

    # Remove query parameters
    route = re.sub(r'\?.*', '', route)

    # Remove image names
    route = re.sub(r'/image/[^/]+', '/image/[id]', route)
    route = re.sub(r'/static/images/[^/]+', '/static/images/[id]', route)

    # Replace numbers with [id]
    route = re.sub(r'\d+', '[id]', route)

    # Remove specific numeric patterns like '150x150'
    route = re.sub(r'\d+x\d+|x\d+|\d+x', '', route)
    
    # Traverse the string and find the first '/' after which a number, '?', or '%' is present
    for i, char in enumerate(route):
        if char == '/' and (any(c.isdigit() for c in route[i:]) or '?' in route[i:] or '%' in route[i:]):
            # Remove everything after the first '/' meeting the condition
            modified_route = route[:i]
            break
    else:
        # If no such '/' is found, keep the original string
        modified_route = route
    
    return modified_route

# Apply the function to the 'request' column
logs_df['request_template'] = logs_df['request'].apply(modify_request)
```

```python
def modify_request(route):
    # Check if the route ends with '.png'
    if route.endswith('.png'):
        # Remove everything after the last '/'
        modified_route = route.rsplit('/', 1)[0]
    else:
        modified_route = route
    
    return modified_route

# Apply the function to the 'request' column
logs_df['request_template'] = logs_df['request_template'].apply(modify_request)
```
